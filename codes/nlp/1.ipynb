## Removing Punctuations and Stop words

```python
from nltk.corpus import stopwords

# nltk.load()
text = "This is a test from rezvan course!"
tokens = word_tokenize(text)
stopword_list = stopwords.words('english')
[token for token in tokens if token not in stopword_list]
```

```python
import string

punc = string.punctuation
[token for token in tokens if token not in punc]
```

## Lemmatization and stemming

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize

text_str = "Artificial intelligence (AI) is a branch of computer science that aims to build machines capable of performing tasks that typically require human intelligence. AI enables machines to simulate human abilities, such as learning, problem-solving, decision-making and comprehension. Common applications of AI include speech recognition, image recognition, content generation, recommendation systems and self-driving cars."
text = word_tokenize(text_str)
lem = WordNetLemmatizer()
lem_ = [lem.lemmatize(token) for token in text]
porterstem = PorterStemmer()
porterstem_ = [porterstem.stem(token) for token in text]
snowstem = SnowballStemmer('english')
snowstem_ = [snowstem.stem(token) for token in text]
' '.join(text), ' '.join(lem_), ' '.join(porterstem_), ' '.join(snowstem_)
```

```python
from textblob import TextBlob
wiki = TextBlob(text_str)
words = wiki.tokenize()
words.lemmatize()
```

```python
from textblob import TextBlob
wiki = TextBlob(text_str)
words = wiki.tokenize()
words.stem()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(text_str)

' '.join([token.lemma_ for token in doc])
```

```python
from hazm import word_tokenize, Lemmatizer, Stemmer

text = 'خدا زمین رو گرد آفرید تا به انسان بگه همون لحظه ای که فکر می کنی به آخر دنیا رسیده ای درست در نقطه آغاز هستی'
tokens = word_tokenize(text)
lemmatizer = Lemmatizer()
' '.join([lemmatizer.lemmatize(token) for token in tokens])
```

## POS Tagging

```python
import spacy

nlp = spacy.load("en_core_web_sm")
text = "This is a test from rezvan course!"
docs = nlp(text)
[token.pos_ for token in docs]
```

```python
import nltk
from nltk import tag
from nltk.tokenize import word_tokenize

nltk.download('averaged_perceptron_tagger_eng')

text = "This is a test from rezvan course!"
words = word_tokenize(text)
tag.pos_tag(words)
```

```python
import hazm

text = 'خدا زمین رو گرد آفرید تا به انسان بگه همون لحظه ای که فکر می کنی به آخر دنیا رسیده ای درست در نقطه آغاز هستی'
words = hazm.word_tokenize(text)
tagger = hazm.POSTagger ( model = 'pos_tagger.model' )
tagger.tag (words)
```
