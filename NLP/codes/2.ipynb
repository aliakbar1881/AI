## Text Modeling

### Navi Bayse

```python
import numpy as np
from tensorflow.keras.datasets import reuters
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=10000)
```

```python
word_index = reuters.get_word_index()
```

```python
index_to_word = {v: k for k, v in word_index.items()}
```

```python
def decode_text(sequence):
    return " ".join([index_to_word.get(i - 3, "?") for i in sequence])

x_train_text = [decode_text(seq) for seq in x_train]
x_test_text = [decode_text(seq) for seq in x_test]
```

```python
vectorizer = CountVectorizer(max_features=5000)
x_train_bow = vectorizer.fit_transform(x_train_text)
x_test_bow = vectorizer.transform(x_test_text)
```

```python
nb_model = MultinomialNB()
nb_model.fit(x_train_bow, y_train)

y_pred = nb_model.predict(x_test_bow)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(x_train_bow, y_train)

y_pred = lr_model.predict(x_test_bow)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

print("Classification Report:")
print(classification_report(y_test, y_pred))
```
