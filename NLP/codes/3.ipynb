### RNN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.datasets import imdb
import numpy as np
```

```python
(train, train_y), (test, test_y) = imdb.load_data(num_words=10000)
```

```python
np.unique(train_y)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences


train = pad_sequences(train, 3000)
test = pad_sequences(test, 3000)
print(train.shape, test.shape)
```

```python

model = Sequential(
    [
        SimpleRNN(64, activation='relu', input_shape=(3000, 1)),
        Dense(1, activation='sigmoid')
    ]
)

model.compile(optimizer='adam', loss='binarycategorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics='accuracy')
```

```python
history = model.fit(
    train,
    train_y,
    epochs=10,
    validation_split=.1,
    batch_size=32,
    shuffle=True
)
```
