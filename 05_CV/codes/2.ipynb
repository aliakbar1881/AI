## Unpooling

```python
from tensorflow.keras.layers import UpSampling2D
import numpy as np
```

```python
x = np.arange(9).reshape((1, 3, 3, 1))
x
```

```python
up = UpSampling2D((2, 2))
```

```python
result = up(x)
```

```python
from pprint import pprint
pprint(result.numpy().reshape(-1, 6))
```
