```python
import numpy as np

# محیط ساده مدیریت انبار
class InventoryEnv:
    def __init__(self, max_inventory=10, max_order=5):
        self.max_inventory = max_inventory
        self.max_order = max_order
        self.reset()

    def reset(self):
        self.inventory = 5
        return self.inventory

    def step(self, action):
        demand = np.random.poisson(3)  # تقاضای تصادفی
        self.inventory = max(0, self.inventory + action - demand)
        cost = 2 * action + 1 * self.inventory  # هزینه سفارش + نگهداری
        done = False
        return self.inventory, -cost, done, {}

env = InventoryEnv()
Q = np.zeros((env.max_inventory + 1, env.max_order + 1))
alpha = 0.1
gamma = 0.95
epsilon = 0.1

for episode in range(5000):
    s = env.reset()
    while True:
        if np.random.rand() < epsilon:
            a = np.random.randint(0, env.max_order + 1)
        else:
            a = np.argmax(Q[s, :])
        s_prime, reward, done, _ = env.step(a)
        Q[s, a] = Q[s, a] + alpha * (reward + gamma * np.max(Q[s_prime, :]) - Q[s, a])
        s = s_prime
        if done:
            break

print("آموزش کامل شد. سیاست بهینه:")
print(np.argmax(Q, axis=1))
```
