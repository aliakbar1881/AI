  <h2 style="color: #2980b9; margin-top: 20px;">Introduction to Pandas and Data Types</h2>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Pandas</strong> is a powerful open-source Python library designed for data manipulation and analysis. Built on top of NumPy, it provides easy-to-use data structures like Series (1D) and DataFrame (2D) to handle structured data efficiently. Whether you're cleaning messy datasets, performing statistical analysis, or preparing data for machine learning, Pandas is your go-to tool. One of its key features is its flexible handling of data types, which allows you to optimize memory usage and ensure accurate computations. Let’s dive into the types of data Pandas supports and how they work!

  <h3 style="color: #2c3e50; margin-top: 15px;">Numeric Data Types</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Pandas (and NumPy, which it builds on) offers a variety of numeric data types for integers and floating-point numbers:
  </p>
  <ul style="margin-left: 20px; color: #1a2526;">
    <li><strong>Integers:</strong>
      <ul>
        <li><code>int8, int16, int32, int64</code>: Signed integers (can be negative or positive) with sizes of 8, 16, 32, or 64 bits.</li>
        <li><code>uint8, uint16, uint32, uint64</code>: Unsigned integers (only positive) with the same bit sizes.</li>
      </ul>
    </li>
    <li><strong>Floating-Point Numbers:</strong>
      <ul>
        <li><code>float16, float32, float64</code>: Floating-point numbers with 16, 32, or 64 bits of precision. <code>float64</code> is also called "double" and is the default in Pandas.</li>
      </ul>
    </li>
  </ul>
  <p style="font-size: 16px; color: #1a2526;">
    <strong>Example:</strong>
    <pre style="background-color:rgb(0, 0, 0); padding: 10px; border-radius: 5px;">
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [1.5, 2.7, 3.2]})
df['A'] = df['A'].astype('int32')  # Convert to 32-bit integer
df['B'] = df['B'].astype('float32')  # Convert to 32-bit float
print(df.dtypes)
# Output:
# A     int32
# B    float32
    </pre>
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">String Data Types</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Strings are handled differently in NumPy and Pandas:
  </p>
  <ul style="margin-left: 20px; color: #1a2526;">
    <li><strong>NumPy:</strong>
      <ul>
        <li><code>dtype='U'</code>: Unicode strings (variable length).</li>
        <li><code>dtype='S'</code>: Byte strings (fixed length).</li>
      </ul>
    </li>
    <li><strong>Pandas:</strong>
      <ul>
        <li><code>object</code>: Default type for strings or mixed data in Pandas.</li>
        <li><code>string</code>: A newer Pandas-specific type (introduced in version 1.0.0) for pure string data, more efficient than <code>object</code>.</li>
      </ul>
    </li>
  </ul>
  <p style="font-size: 16px; color: #1a2526;">
    <strong>Example:</strong>
    <pre style="background-color:rgb(0, 0, 0); padding: 10px; border-radius: 5px;">
df = pd.DataFrame({'Names': ['Ali', 'Sara', 'Omid']})
print(df.dtypes)  # Names: object
df['Names'] = df['Names'].astype('string')  # Convert to string type
print(df.dtypes)  # Names: string
    </pre>
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">Boolean Data Type</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Both Pandas and NumPy support a boolean type:
  </p>
  <ul style="margin-left: 20px; color: #1a2526;">
    <li><code>bool</code>: Represents <code>True</code> or <code>False</code> values.</li>
  </ul>
  <p style="font-size: 16px; color: #1a2526;">
    <strong>Example:</strong>
    <pre style="background-color:rgb(0, 0, 0); padding: 10px; border-radius: 5px;">
df = pd.DataFrame({'Active': [True, False, True]})
print(df.dtypes)  # Active: bool
    </pre>
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">DateTime Data Types</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Handling dates and times is essential for time-series analysis:
  </p>
  <ul style="margin-left: 20px; color: #1a2526;">
    <li><strong>NumPy:</strong>
      <ul>
        <li><code>datetime64</code>: General type for dates and times.</li>
      </ul>
    </li>
    <li><strong>Pandas:</strong>
      <ul>
        <li><code>datetime64[ns]</code>: High-precision datetime with nanosecond resolution.</li>
        <li><code>timedelta64[ns]</code>: For differences between times (e.g., durations).</li>
      </ul>
    </li>
  </ul>
  <p style="font-size: 16px; color: #1a2526;">
    <strong>Example:</strong>
    <pre style="background-color:rgb(0, 0, 0); padding: 10px; border-radius: 5px;">
df = pd.DataFrame({'Dates': ['2023-01-01', '2023-01-02']})
df['Dates'] = pd.to_datetime(df['Dates'])  # Convert to datetime64[ns]
print(df.dtypes)  # Dates: datetime64[ns]
df['Diff'] = df['Dates'] - pd.to_datetime('2023-01-01')
print(df.dtypes)  # Diff: timedelta64[ns]
    </pre>
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">Categorical Data Type</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Pandas provides a special type for categorical data, which is great for repeated values:
  </p>
  <ul style="margin-left: 20px; color: #1a2526;">
    <li><code>category</code>: Stores unique values efficiently and is ideal for columns with limited, repeated entries (e.g., gender, days of the week).</li>
  </ul>
  <p style="font-size: 16px; color: #1a2526;">
    <strong>Example:</strong>
    <pre style="background-color:rgb(0, 0, 0); padding: 10px; border-radius: 5px;">
df = pd.DataFrame({'Day': ['Mon', 'Tue', 'Mon', 'Wed']})
df['Day'] = df['Day'].astype('category')
print(df.dtypes)  # Day: category
print(df['Day'].cat.categories)  # Index(['Mon', 'Tue', 'Wed'], dtype='object')
    </pre>
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">Special Data Types</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Some additional types for specific use cases:
  </p>
  <ul style="margin-left: 20px; color: #1a2526;">
    <li><strong>NumPy:</strong>
      <ul>
        <li><code>complex</code>: For complex numbers (e.g., 3 + 4j).</li>
      </ul>
    </li>
    <li><strong>Pandas:</strong>
      <ul>
        <li><code>Int64, Float64</code>: Nullable integer and float types that can handle <code>NaN</code> without converting to float.</li>
        <li><code>sparse</code>: For datasets with lots of zeros, saving memory.</li>
      </ul>
    </li>
  </ul>
  <p style="font-size: 16px; color: #1a2526;">
    <strong>Example:</strong>
    <pre style="background-color:rgb(0, 0, 0); padding: 10px; border-radius: 5px;">
df = pd.DataFrame({'A': [1, None, 3]})
df['A'] = df['A'].astype('Int64')  # Nullable integer
print(df.dtypes)  # A: Int64
print(df)  # Can keep NaN without changing type
    </pre>
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">The <code>dtype</code> Parameter</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    The <code>dtype</code> parameter in Pandas (and NumPy) lets you specify the data type when creating or converting data. It’s super useful for controlling memory usage and ensuring your data behaves as expected. You can pass it to functions like <code>pd.DataFrame()</code>, <code>pd.Series()</code>, or <code>astype()</code>.
  </p>
  <p style="font-size: 16px; color: #1a2526;">
    <strong>Example:</strong>
    <pre style="background-color:rgb(0, 0, 0); padding: 10px; border-radius: 5px;">
# Creating a DataFrame with specific dtypes
df = pd.DataFrame({'A': [1, 2, 3], 'B': ['x', 'y', 'z']}, dtype={'A': 'int8', 'B': 'string'})
print(df.dtypes)  # A: int8, B: string

# Converting with astype
df['A'] = df['A'].astype('float32')
print(df.dtypes)  # A: float32, B: string
    </pre>

  <h1 style="color: #2c3e50; text-align: center; border-bottom: 2px solid #3498db; padding-bottom: 10px;">Create a Test Set</h1>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Before exploring further, it’s critical to set aside a test set (and sometimes a validation set) to evaluate your model later. This prevents "data snooping" bias—where you unintentionally overfit to the entire dataset. Here’s how to split data:
  </p>
  <img src="https://algotrading101.com/learn/wp-content/uploads/2020/06/training-validation-test-data-set.png" 
alt="dataframe and series" style="max-width: 100%; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>1.1 Basic Train-Test Split:</strong><br>
    Use <code>train_test_split</code> from <code>sklearn.model_selection</code>:
    <ul style="margin-left: 20px; color: #1a2526;">
      <li>Splits data into training (e.g., 80%) and test (e.g., 20%) sets.</li>
      <li><code>random_state</code> ensures reproducibility.</li>
    </ul>
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>1.2 Train-Validation-Test Split:</strong><br>
    For tuning models, split data into three parts:
    <ul style="margin-left: 20px; color: #1a2526;">
      <li>Training (e.g., 60%): For model learning.</li>
      <li>Validation (e.g., 20%): For tuning hyperparameters.</li>
      <li>Test (e.g., 20%): For final evaluation.</li>
    </ul>
    Can be done with multiple <code>train_test_split</code> calls.
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>1.3 StratifiedShuffleSplit</strong> is a powerful tool from <code>sklearn.model_selection</code> that combines stratified sampling with shuffling to create multiple train-test splits. Unlike a single <code>train_test_split</code>, it generates several splits while ensuring that the proportion of categories (e.g., classes or groups) remains consistent across all splits. This is especially useful for cross-validation, evaluating model stability, or when you need repeated stratified samples from an imbalanced dataset.
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Key Points:</strong>
    <ul style="margin-left: 20px; color: #1a2526;">
      <li><strong>When to Use:</strong> For cross-validation with imbalanced data or when you want multiple stratified splits.</li>
      <li><strong>How It Works:</strong> It shuffles the data, then splits it into training and test sets multiple times, maintaining category proportions based on a specified column.</li>
      <li><strong>Parameters:</strong>
        <ul>
          <li><code>n_splits</code>: Number of re-shuffling & splitting iterations.</li>
          <li><code>test_size</code>: Proportion of data for the test set.</li>
          <li><code>train_size</code>: Proportion for the training set (optional).</li>
          <li><code>random_state</code>: For reproducibility.</li>
        </ul>
      </li>
    </ul>
  </p>
</div>

```python
from sklearn.model_selection import train_test_split

# Basic Train-Test Split
train_set, test_set = train_test_split(combined_df, test_size=0.2, random_state=42)
print("Training set size:", len(train_set))
print("Test set size:", len(test_set))

# Train-Validation-Test Split
train_val, test_set = train_test_split(combined_df, test_size=0.2, random_state=42)
train_set, val_set = train_test_split(train_val, test_size=0.25, random_state=42)  # 0.25 of 80% = 20% of original
print("\nTrain set size:", len(train_set))
print("Validation set size:", len(val_set))
print("Test set size:", len(test_set))
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
import seaborn as sns
import pandas as pd

# Load the tips dataset
tips = sns.load_dataset("tips")

# Initialize StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42) # n_split=Number of re-shuffling & splitting iterations

# Use 'day' column for stratification
X = tips.drop(columns=['day'])  # Features (could be all columns except the stratification column)
y = tips['day']  # Stratification target

# Iterate over splits
for train_index , test_index in split.split(X, y):
    strat_train_set = tips.iloc[train_index]
    strat_test_set = tips.iloc[test_index]
    print("\nTraining set 'day' distribution:\n", strat_train_set['day'].value_counts(normalize=True))
    print("Test set 'day' distribution:\n", strat_test_set['day'].value_counts(normalize=True))
```

  <h2 style="color: #2980b9; margin-top: 20px;">5. Key Components of EDA</h2>

```python
def clean_stops(stop_value):
    stop_value = str(stop_value).replace('\n', '').replace('\t', '').strip()
    stop_value = stop_value.split('Via')[0].strip()
    if 'non-stop' in stop_value.lower():
        return 'zero'
    elif '1-stop' in stop_value:
        return 'one'
    elif '2+' in stop_value:
        return 'two_or_more'
    else:
        return stop_value

train_set['stop'] = train_set['stop'].apply(clean_stops)
train_set['stop'].value_counts()
```

## Normal and Standard :

<div style="font-family: Arial, sans-serif; max-width: 1000px; margin: 0 auto; padding: 20px; line-height: 1.6; background-color: #f9f9f9; border-radius: 10px; box-shadow: 0 4px 8px rgba(0,0,0,0.1);">
  <h2 style="color: #2980b9; margin-top: 20px;">6. What is a Normal Distribution?</h2>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    The <strong>Normal Distribution</strong> is a cornerstone of statistics and data science, often appearing as a bell-shaped curve. It describes how data points are spread around a central value (the mean), with most values clustering near the mean and fewer as you move away in either direction. This distribution is key in many real-world scenarios, like test scores or measurement errors.
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Key Features of Normal Distribution:</strong>
    <ul style="margin-left: 20px; color: #1a2526;">
      <li><strong>Symmetry:</strong> The curve is perfectly symmetric around the mean.</li>
      <li><strong>Mean, Median, Mode:</strong> All are equal and located at the center.</li>
      <ul style="margin-left: 20px; color: #1a2526;">
        <li><strong>Mean:</strong> The average of all data points, calculated as the sum divided by the count (μ = Σx / n).</li>
        <li><strong>Median:</strong> The middle value when data is sorted, splitting the dataset into two equal halves.</li>
        <li><strong>Mode:</strong> The most frequently occurring value in the dataset.</li>
      </ul>
      <li><strong>Spread:</strong> Controlled by the standard deviation (σ)—a larger σ means a wider curve.</li>
    </ul>
    Mathematically, it’s defined by the probability density function (PDF): 
    <code>f(x) = (1 / (σ√(2π))) * e^(-(x-μ)²/(2σ²))</code>, where μ is the mean and σ is the standard deviation.
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">6.1 Normal vs. Standard Normal Distribution</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    A <strong>Standard Normal Distribution</strong> is a special case of the normal distribution with a mean (μ) of 0 and a standard deviation (σ) of 1. Any normal distribution can be transformed into a standard normal one using the z-score formula: <code>z = (x - μ) / σ</code>. This standardization simplifies calculations and comparisons.
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Differences:</strong>
    <ul style="margin-left: 20px; color: #1a2526;">
      <li><strong>Mean and Spread:</strong> Normal distribution can have any mean and standard deviation; standard normal is fixed at μ = 0, σ = 1.</li>
      <li><strong>Use Case:</strong> Standard normal is used for statistical tests (e.g., z-tests) and probability tables.</li>
    </ul>
    <strong>💡 Tip:</strong> Understanding these distributions helps in data normalization and assumption checking for many machine learning models!
  </p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data
x = np.linspace(-10, 10, 1000)  # Range of x values

# Normal distribution (μ = 2, σ = 1.5)
normal_dist = norm.pdf(x, loc=2, scale=1.5)

# Standard normal distribution (μ = 0, σ = 1)
std_normal_dist = norm.pdf(x, loc=0, scale=1)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, normal_dist, label='Normal (μ=2, σ=1.5)', color='blue')
plt.plot(x, std_normal_dist, label='Standard Normal (μ=0, σ=1)', color='red', linestyle='--')
plt.title('Normal vs. Standard Normal Distribution')
plt.xlabel('X Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

  <h2 style="color: #2980b9; margin-top: 20px;">Understanding Skewness and Kurtosis</h2>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    When exploring data, it’s not enough to just look at averages or spread. The <strong>shape</strong> of the data distribution matters too! Two key measures, <strong>Skewness</strong> and <strong>Kurtosis</strong>, help us understand this shape better. Let’s break them down.
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">Skewness</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Skewness measures the <strong>symmetry</strong> (or lack thereof) in a dataset’s distribution. It tells us whether the data leans more to the left or right.
    <ul style="margin-left: 20px; color: #1a2526;">
      <li><strong>Zero Skewness (Symmetric):</strong> The data is perfectly balanced, like a bell curve (normal distribution). The left and right tails are equal.</li>
      <li><strong>Positive Skewness (Right-Skewed):</strong> The right tail is longer or fatter—data piles up on the left, with some larger values stretching out to the right (e.g., income distributions).</li>
      <li><strong>Negative Skewness (Left-Skewed):</strong> The left tail is longer or fatter—data piles up on the right, with some smaller values stretching out to the left (e.g., time to failure of mechanical systems).</li>
    </ul>
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Why It Matters:</strong> Skewness affects how we preprocess data (e.g., applying log transformations for right-skewed data) and choose models, since many assume normality.
  </p>

  <h3 style="color: #2c3e50; margin-top: 15px;">Kurtosis</h3>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    Kurtosis measures the <strong>tailedness</strong> of a distribution—how heavy or light the tails are compared to a normal distribution. It tells us about the presence of outliers and the “peakiness” of the data.
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Why It Matters:</strong> High kurtosis might signal potential outliers that need handling, while low kurtosis suggests a more even spread, impacting statistical tests or model assumptions.
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>💡 Pro Tip:</strong> Use skewness and kurtosis together to get a fuller picture of your data’s shape—symmetry from skewness, and tail behavior from kurtosis.
  </p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm, skewnorm, laplace

# Set random seed for reproducibility
np.random.seed(42)

# Generate sample datasets with different skewness and kurtosis
normal_data = norm.rvs(size=1000, loc=0, scale=1)  # Normal (symmetric, mesokurtic)
right_skewed = skewnorm.rvs(a=5, size=1000)  # Right-skewed (positive skewness)
left_skewed = skewnorm.rvs(a=-5, size=1000)  # Left-skewed (negative skewness)
high_kurtosis = laplace.rvs(size=1000, loc=0, scale=1)  # Leptokurtic (high kurtosis)

# Create subplots
plt.figure(figsize=(12, 8))

# Normal Distribution
plt.subplot(2, 2, 1)
sns.histplot(normal_data, kde=True, color='blue', stat='density')
plt.title(f'Normal\nSkewness: {skew(normal_data):.2f}, Kurtosis: {kurtosis(normal_data):.2f}')
plt.xlabel('Value')

# Right-Skewed Distribution
plt.subplot(2, 2, 2)
sns.histplot(right_skewed, kde=True, color='green', stat='density')
plt.title(f'Right-Skewed\nSkewness: {skew(right_skewed):.2f}, Kurtosis: {kurtosis(right_skewed):.2f}')
plt.xlabel('Value')

# Left-Skewed Distribution
plt.subplot(2, 2, 3)
sns.histplot(left_skewed, kde=True, color='red', stat='density')
plt.title(f'Left-Skewed\nSkewness: {skew(left_skewed):.2f}, Kurtosis: {kurtosis(left_skewed):.2f}')
plt.xlabel('Value')

# High Kurtosis Distribution
plt.subplot(2, 2, 4)
sns.histplot(high_kurtosis, kde=True, color='purple', stat='density')
plt.title(f'High Kurtosis\nSkewness: {skew(high_kurtosis):.2f}, Kurtosis: {kurtosis(high_kurtosis):.2f}')
plt.xlabel('Value')

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
# Apply log transformation to right-skewed data (shift to ensure all values > 0)
# Adding 1 - min to avoid log(0) or negative values
right_skewed_shifted = right_skewed - right_skewed.min() + 1  # Shift to positive
right_skewed_log = np.log(right_skewed_shifted)  # Log transformation
# Right-Skewed with Log Transformation
sns.histplot(right_skewed_log, kde=True, color='orange', stat='density')
plt.title(f'Right-Skewed (Log Transformed)\nSkewness: {skew(right_skewed_log):.2f}, Kurtosis: {kurtosis(right_skewed_log):.2f}')
plt.xlabel('Log(Value)')
plt.show()
```

  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Box Plots and Quartiles:</strong> Box plots visualize data distribution using quartiles and help identify outliers.
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>What are Quartiles and IQR?</strong><br>
    <ul style="margin-left: 20px; color: #1a2526;">
      <li><strong>Quartiles:</strong> Values dividing data into four equal parts.</li>
        <ul>
          <li><strong>Q1 (First Quartile):</strong> 25th percentile (25% of data below).</li>
          <li><strong>Q2 (Median):</strong> 50th percentile (50% below).</li>
          <li><strong>Q3 (Third Quartile):</strong> 75th percentile (75% below).</li>
        </ul>
      <li><strong>IQR (Interquartile Range):</strong> Spread of the middle 50% of data, calculated as <code>IQR = Q3 - Q1</code>.</li>
    </ul>
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>How to Draw a Box Plot:</strong>
    <ol style="margin-left: 20px; color: #1a2526;">
      <li>Order the data in ascending order.</li>
      <li>Calculate Q1, Q2 (median), and Q3.</li>
      <li>Identify min and max values (or within 1.5 × IQR for outliers).</li>
      <li>Draw the box from Q1 to Q3, with a line at Q2.</li>
      <li>Extend whiskers to min/max or outlier fences.</li>
    </ol>
    <strong>Outlier Detection:</strong><br>
    - Lower Fence: <code>Q1 - (1.5 × IQR)</code><br>
    - Upper Fence: <code>Q3 + (1.5 × IQR)</code>
  </p>
  <p style="font-size: 16px; color: #1a2526; text-align: justify;">
    <strong>Example Dataset:</strong> [6, 12, 14, 16, 20, 21, 22, 24, 28, 36, 40]<br>
    - Min: 6, Q1: 14, Q2: 21, Q3: 28, Max: 40<br>
    Draw a box from 14 to 28, line at 21, whiskers to 6 and 40.
  </p>
