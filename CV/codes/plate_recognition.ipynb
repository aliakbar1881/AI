## Plate Recognition Project

> dataset: https://github.com/mut-deep/IR-LPR

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data():
    ds_dir = './dataset/'
    ds = image_dataset_from_directory(ds_dir)
    return ds

ds = load_data()
```

```python
for image, label in ds.take(1):
    plt.imshow(image[0].numpy().astype(int))
    plt.axis('off')
    plt.show()
    plt.imshow(image[10].numpy().astype(int))
    plt.axis('off')
    plt.show()
    plt.imshow(image[20].numpy().astype(int))
    plt.axis('off')
```

```python
def augment(image, label):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
    return image, label

ds = ds.map(augment)
ds = ds.prefetch(tf.data.AUTOTUNE)
```

```python
for image, label in ds.take(1):
    plt.imshow(image[0].numpy().astype(int))
    plt.axis('off')
    plt.show()
    plt.imshow(image[10].numpy().astype(int))
    plt.axis('off')
    plt.show()
    plt.imshow(image[20].numpy().astype(int))
    plt.axis('off')
```
